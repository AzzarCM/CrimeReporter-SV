```python
#Importando datos de CSV
df = pd.read_csv("drive/My Drive/IMLHomCSVPrueba.csv")
print(len(df))
df.head()
```

```python
import datetime
import time as timelib

timestamp = []

def fourDigitsTime(time):
  try:
    intHora = int(time)

    if(len(time) == 1):
      strTime = '000' + time
    elif(len(time) == 2): 
      strTime = '00' + time
    elif(len(time) == 3): 
      strTime = '0' + time
    else: 
      strTime = time
  except:
    strTime = '10000'
  
  return strTime

def twoDigitsDate(date):
  return '0' + date if(len(date) == 1) else date

for date, time, row in zip(df["Fecha_Completa"], df["Hora"], df["NoTotal"]):

  #Formateando fecha
  splitDate = date.split('/')
  month = twoDigitsDate(splitDate[0])
  day = twoDigitsDate(splitDate[0])
  year = splitDate[2]
  formattedDated = year + '-' + month + '-' + day

  #Formateando hora
  newTime = fourDigitsTime(time.strip())
  timeList = list(newTime)
  hour = timeList[0] + timeList[1]
  minutes = timeList[2] + timeList[3]
  formattedTime = hour + ':' + minutes+ ':'+'00'
  intTime = int(hour+minutes)
  if(int(hour) > 23 or int(minutes) > 59):

    df.drop(df[df.NoTotal == row].index, inplace=True)
  else:
    ts = datetime.datetime.strptime(formattedDated+' '+formattedTime, '%Y-%m-%d %H:%M:%S')
    timestamp.append(timelib.mktime(ts.timetuple()))

print(timestamp)
```

```python
#Agrega la columna de tiempo
timeStamp = pd.Series(timestamp)
df['Timestamp'] = timeStamp.values
df.head()
```

```python
#Elimina columnas innecesarias para el data set
data_set1 = df.drop(['NoTotal', 'Año', 'N', 'Mes', 'Fecha', 'Fecha_Completa', 'Dia','Hora', 'Rango_Horario', 'Departamento', 'Codigo_Departamento', 'Codigo_Municipio'], axis=1)
data_set2 = data_set1.drop(['Rango_Edad', 'Clasificacion_Arma'], axis=1)
```
